# Data Manipulation #

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Name': ['Ram', 'Mohan', 'Rani','Shyam', None, 'Raj'],
    'Age': [25, None,40,21,36,25],
    'Salary': [50000, 40000,60000,None,80000, 70000]
})
```

```python
#Handling missing values
df['Name'].fillna('Unknown', inplace=True)
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)

df
```

```python
# Add new column
df['Yearly_Salary'] =  df['Salary'] * 12

df
```

```python
# Rename colum name
df.rename(columns={'Salary': 'Monthly Salary' })
```

```python
#  Filter Data
filter_salary =  df[df['Yearly_Salary'] >= 720000]
filter_salary
```
