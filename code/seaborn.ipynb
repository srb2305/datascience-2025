```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#sample data

df = pd.DataFrame({
    "Day" : ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Sales": [6000, 7000,8000,2000,1000,10000,16000]
})

plt.plot(df["Day"], df["Sales"], color="blue", marker="o")
plt.xlabel("Day x-axis")
plt.ylabel("Sales y-axis")
plt.title("Matplotlib Sales Data Day wise")
plt.show()

sns.lineplot(x="Day", y="Sales", data=df, marker="o", color="blue")
plt.title("SNS Sales Data Day wise")
plt.show()
```

```python
sns.lineplot(x="Day", y="Sales", data=df, 
             marker="o",
             color="blue",
             markerfacecolor="red",
             markeredgecolor="black",
             linestyle="--"
            )
plt.title("SNS Sales Data Day wise")
plt.show()
```

```python
sns.barplot(x="Day", y="Sales", data=df, 
             color="orange",
             palette="muted"
            )
plt.title("SNS Sales Data Day wise")
plt.show()
```

```python
sns.scatterplot(x="Day", y="Sales", data=df, 
               color="orange",
                s=80,
                zorder=5
            )

sns.lineplot(x="Day", y="Sales", data=df, 
               color="red"
            )

sns.barplot(x="Day", y="Sales", data=df, 
               color="lightblue"
            )

plt.title("SNS Sales Data Day wise")
plt.grid(1)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y, color="blue", marker="o", label="Sin")
sns.lineplot(x=x,y=z, color="orange", marker="o", label="Cos")

plt.grid(1)
#plt.legend()
plt.show()
```
