# Matplotlib #
 **Purpose: Data visualization library**

 - Why use in data science
     - Create line charts, bar,  scatter plots, histograms etc.
     - Work directly with numpy and pandas

```python
import matplotlib.pyplot as plt

x = [ 1,2,3,4 ]
y = [ 3,4,7,8 ]

plt.plot(x, y, marker="s", color="orange")   # Line chart    marker = o, ., d, D, X, x, h, H, s 
plt.title(" Simple Plot ")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data  = {
    'Name' : ['Ram', 'Shyam', 'Mohan', 'Sohan', 'Seeta','Geeta','Reeta','RajaRam','RajaMohan'],
    'Score': [100,50,20,60, 10,50,60,80,90]
}

df = pd.DataFrame(data)
#plt.bar(df['Name'], df['Score'], color="red")
#plt.barh(df['Name'], df['Score'], color="red")
#plt.scatter(df['Name'], df['Score'], color="red")
plt.fill_between(df['Name'], df['Score'], color="red")

plt.title('Students Grade')
plt.xlabel('Student Name')
plt.ylabel('Students Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
